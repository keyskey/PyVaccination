```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = 16, 7
plt.rcParams['font.size'] = 14
df = pd.read_csv('episode0.csv') 

pivot1 = df.pivot('Effectiveness', 'Cr', 'FES')
pivot2 = df.pivot('Effectiveness', 'Cr', 'VC')
pivot3 = df.pivot('Effectiveness', 'Cr', 'SAP')

fig = plt.figure()

ax1 = fig.add_subplot(131)
sns.heatmap(pivot1, cmap='Reds', vmin = 0, vmax = 1.0, square=True, annot=False, cbar_kws={"orientation": "horizontal"})
ax1.set_title('FES')
plt.gca().invert_yaxis()

ax2 = fig.add_subplot(132)
sns.heatmap(pivot2, cmap='Blues', vmin = 0, vmax = 1.0, square=True, annot=False, cbar_kws={"orientation": "horizontal"})
ax2.set_title('VC')
plt.gca().invert_yaxis()

ax3 = fig.add_subplot(133)
sns.heatmap(pivot3, cmap='YlGnBu', vmin = -1, vmax= 0, square=True, annot=False, cbar_kws={"orientation": "horizontal"})
ax3.set_title('SAP')
plt.gca().invert_yaxis()

plt.tight_layout()
plt.savefig("episode0.png")
plt.show()
```
